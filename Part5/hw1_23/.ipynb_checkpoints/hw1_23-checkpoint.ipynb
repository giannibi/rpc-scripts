```python
import matplotlib.pyplot as plt
from itertools import product
import slycot as sly
import numpy as np
import control as ct
import rpc_diism.robust as rb
s= ct.tf('s')
```

```python
def step_opposite(g, t):
    # Compute reponse to step of [1,-1] of a system with 2 inputs and 2 outputs
    _, yu1 = ct.step_response(g, t, input=0, squeeze=True)
    _, yu2 = ct.step_response(g, t, input=1, squeeze=True)
    return yu1 - yu2
    

def plot_performance(w, P, K, S, We, Wu, gamma):
    
    I = ct.ss([], [], [], np.eye(2)) # Identity system
        
    # Compute singular values of S
    Ssv = rb.sigma(S, w)
    
    # Compute singular values of nubar*inv(We)
    Weinv = rb.invss(ct.ss(We))
    Weinvsv = rb.sigma(Weinv, w) * gamma
    
    # Compute input sensitivity
    KS = K*S
    
    # Compute singular values of KS
    KSsv = rb.sigma(KS, w)
    
    # Compute singular values of nubar*inv(Wu)
    Wuinv = rb.invss(ct.ss(Wu))
    Wuinvsv = rb.sigma(Wuinv, w) * gamma
    

    plt.figure(figsize=(15, 6))
    
    # Plot singular values of S
    plt.subplot(1, 3, 1)
    plt.semilogx(w, 20*np.log10(Ssv[:, 0]), 'm', label=r'$\sigma_1(S)$')
    plt.semilogx(w, 20*np.log10(Ssv[:, 1]), 'g', label=r'$\sigma_2(S)$')
    plt.semilogx(w, 20*np.log10(Weinvsv[:, 0]), 'm:', label=r'$\bar\nu \sigma_1({Ws_1}^{-1})$')
    plt.semilogx(w, 20*np.log10(Weinvsv[:, 1]), 'g:', label=r'$\bar\nu \sigma_2({Ws_1}^{-1})$')
    plt.ylabel('magnitude [dB]')
    plt.xlim([1e-5, 1e3])
    plt.xlabel('freq [rad/s]')
    plt.legend()
    plt.title('Singular values of S vs. nubar*inv(We)')
    plt.fill_between(w, -20, -200, where=(w <= 0.1), color='c', alpha=0.2)
    plt.ylim([-70, 20])
    
    # Plot singular values of KS
    plt.subplot(1, 3, 2)
    plt.semilogx(w, 20*np.log10(KSsv[:, 0]), 'm', label=r'$\sigma_1(KS)$')
    plt.semilogx(w, 20*np.log10(KSsv[:, 1]), 'g', label=r'$\sigma_2(KS)$')
    plt.semilogx(w, 20*np.log10(Wuinvsv[:, 0]), 'm:', label=r'$\bar\nu \sigma_1({Wu}^{-1})$')
    plt.semilogx(w, 20*np.log10(Wuinvsv[:, 1]), 'g:', label=r'$\bar\nu \sigma_2({Wu}^{-1})$')
    plt.ylabel('magnitude [dB]')
    plt.xlim([1e-5, 1e3])
    plt.xlabel('freq [rad/s]')
    plt.legend()
    plt.title('Singular values of KS vs. nubar*inv(Wu)')
    plt.fill_between(w, -10, -200, where=(w > 200), color='c', alpha=0.2)
    plt.ylim([-70, 30])
        
    # Plot tracking error
    plt.subplot(1, 3, 3)
    t = np.linspace(0, 10, 101)
    
    T = (ct.ss(P*K)).feedback(I)
    
    y = step_opposite(T,t)
    c = np.round(y[0][-1],4)
    d = np.round(y[1][-1],4)
    
    plt.plot(t, 1-y[0], 'b', label= f' $ y_1$ at steady state = {c}' )
    plt.plot(t, -1-y[1], 'r', label= f' $ y_2$ at steady state = {d}')
    plt.xlabel('time')
    plt.ylabel('outputs')
    plt.legend()
    plt.title('Tracking error to reference [1,-1]')
    plt.xlim([0, 10])
    plt.tight_layout()
    plt.show()
```

```python
def plot_monte_carlo(Kr, We, Wu, w, taumin1, taumax1, taumin2, taumax2, gainmin1, gainmax1, gainmin2, gainmax2, best_nubar):

    plt.subplots(1,3,figsize=(15, 6))
    
    # Generate some random systems in the uncertainty set
    C = []
    D = []
    omega = w
    for i in range(100):
        t_1 = np.round(np.random.uniform(taumin1,taumax1),5)
        t_2 = np.round(np.random.uniform(taumin2,taumax2),5)
        k_1 = np.round(np.random.uniform(gainmin1,gainmax1),5)
        k_2 = np.round(np.random.uniform(gainmin2,gainmax2),5)
        
        Pp = ct.tf([[[k_1],[-0.05]],
               [[0.1],[-k_2]]],
              [[[t_1, 1],[0.1, 1]],
               [[0.3, 1],[t_2, 1]]])
    
        # Compute sensitivities of closed loop system
        I = ct.ss([], [], [], np.eye(2)) # Identity system
        S_ = ct.feedback(I,Pp*Kr)
    
        # Compute singular values of S
        S_sv = rb.sigma(S_, w)
        # Compute singular values of KS
        KS_sv = rb.sigma(Kr*S_, w) 
    
        T = (ct.ss(Pp*Kr)).feedback(I)
    
        t = np.linspace(0, 20, 101)
        y = step_opposite(T,t)
        c = np.round(y[0][-1],4)
        d = np.round(y[1][-1],4)
    
        C.append(c)
        D.append(d)
        
        plt.subplot(1, 3, 1)
        plt.semilogx(w, 20*np.log10(S_sv[:, 0]), 'm')
        plt.semilogx(w, 20*np.log10(S_sv[:, 1]), 'g')
    
        plt.subplot(1, 3, 2)
        plt.semilogx(w, 20*np.log10(KS_sv[:, 0]), 'm')
        plt.semilogx(w, 20*np.log10(KS_sv[:, 1]), 'g')
    
        plt.subplot(1, 3, 3)
        plt.plot(t, 1-y[0], 'b')
        plt.plot(t, -1-y[1], 'r')
    
    Wuinv = rb.invss(ct.ss(Wu))
    Wuinvsv = rb.sigma(Wuinv, omega) * best_nubar
    Weinv = rb.invss(ct.ss(We))  
    Weinvsv = rb.sigma(Weinv, omega) * best_nubar
    
    
    # Plot singular values of S for all systems
    plt.subplot(1, 3, 1)
    plt.fill_between(w, -20, -200, where=(w <= 0.1), color='c', alpha=0.2)
    plt.semilogx(w, 20*np.log10(Weinvsv[:, 0]), 'm:')
    plt.semilogx(w, 20*np.log10(Weinvsv[:, 1]), 'g:')
    plt.title('Singular values of S')
    plt.ylim([-60, 10])
    plt.ylabel('magnitude [dB]')
    plt.xlim([1e-3, 1e3])
    plt.xlabel('freq [rad/s]')

     # Plot singular values of KS for all systems
    plt.subplot(1, 3, 2)
    plt.fill_between(w, -10, -200, where=(w > 200), color='c', alpha=0.2)
    plt.ylim([-70, 20])
    plt.semilogx(w, 20*np.log10(Wuinvsv[:, 0]), 'm:')
    plt.semilogx(w, 20*np.log10(Wuinvsv[:, 1]), 'g:')
    plt.title('Singular values of KS')
    plt.ylabel('magnitude [dB]')
    plt.xlim([1e-3, 1e3])
    plt.xlabel('freq [rad/s]')
    
    # Plot tracking error for all systems
    plt.subplot(1, 3, 3)
    plt.xlabel('time')
    plt.ylabel('outputs')
    
    plt.title('Tracking error to reference [1,-1]')
    #plt.xlim([0, 10])
    plt.tight_layout()
    plt.show()
    
    # Verify that at the end of the simulation time, no system has a tracking error over the limit
    male = False
    for i in range(len(C)):
        if C[i] >= 1-1e-3 and C[i] <= 1+1e-3 and D[i] >= -1-1e-3 and D[i] <= -1+1e-3:
            j = 0
        else:
            male = True
    if not male:
        print('No system exceeds steady state error of 0.001')
    else:
        print('-- Max steady state error exceeded --')
```

Main program. Generate the required blocks for uncertain plant representation. The uncertainty block has the structure $\Delta = blkdiag(\Delta_1,\Delta_2,\Delta_3,\Delta_4)$ where each $\Delta_i$ is scalar

```python
k10=12
k20=5
k1max=1.15*k10
k1min=0.85*k10
k2max=1.15*k20
k2min=0.85*k20
T10=0.2
T20=0.7
T1max=T10*1.2
T1min=T10*0.8
T2max=T20*1.2
T2min=T20*0.8
alpha1 = (k1max-k1min)/2/k10
alpha3 = (k2max-k2min)/2/k20
alpha2 = (T1max-T1min)/2/T10
alpha4 = (T2max-T2min)/2/T20

P21=ct.tf([0.1],[0.3,1],inputs='u[0]',outputs='y21')
P12=ct.tf([-0.05],[0.1,1],inputs='u[1]',outputs='y12')

K1 = ct.ss([],[],[],k10, inputs='u[0]',outputs='q1')
K2 = ct.ss([],[],[],k20, inputs='u[1]',outputs='q3')
A1 = ct.ss([],[],[],alpha1, inputs='q1',outputs='zdelta[0]')
A3 = ct.ss([],[],[],alpha3, inputs='q3',outputs='zdelta[2]')

W2 = ct.tf([alpha2*T10,0.0],[T10,1], inputs='y11',outputs='zdelta[1]')
W4 = ct.tf([alpha4*T20,0.0],[T20,1], inputs='y22',outputs='zdelta[3]')

sump2 = ct.summing_junction(inputs=['wdelta[0]','q1'], output='p2')
sump4 = ct.summing_junction(inputs=['wdelta[2]','q3'], output='p4')

P11=ct.tf([1],[T10,1],inputs='p2',outputs='q2')
P22=ct.tf([-1],[T20,1],inputs='p4',outputs='q4')

sumy11 = ct.summing_junction(inputs=['q2','wdelta[1]'], output='y11')
sumy22 = ct.summing_junction(inputs=['q4','wdelta[3]'], output='y22')

sumy1 = ct.summing_junction(inputs=['y11','y12'], output='y[0]')
sumy2 = ct.summing_junction(inputs=['y21','y22'], output='y[1]')

sumnoise = ct.summing_junction(inputs=['y','n'], output='ybar', dimension=2)

# Feedback node
fbk = ct.summing_junction(inputs=['r','-ybar'], output='e', dimension=2)
```

Performance weights

```python
# Performance weight We (weight on S)
M = 3
e = 1e-4
wb = 1 # Crossover frequency
We_1 = ct.ss(rb.weightS(wb, M, e, 1))
We = We_1.append(We_1)
We.input_labels = ['e[0]', 'e[1]']
We.output_labels = ['z[0]', 'z[1]']

# Weight on command input (weight on KS)
Mu = 0.62
eu = 1e-4
wbc = 2*np.pi*10 # Crossover frequency
Wu11 = ct.ss(rb.weightU(wbc, Mu, eu, 1))
Wu = Wu11.append(Wu11)
Wu.input_labels = ['u[0]', 'u[1]']
Wu.output_labels = ['z[2]', 'z[3]']

omega = np.logspace(-5, 3, 121)

# Compute singular values of inv(We)
Weinv = rb.invss(ct.ss(We))
Weinvsv = rb.sigma(Weinv, omega)

# Compute singular values of inv(Wu)
Wuinv = rb.invss(ct.ss(Wu))
Wuinvsv = rb.sigma(Wuinv, omega)

plt.figure(figsize=(15, 6))
# Plot singular values of inv(We)
plt.semilogx(omega, 20*np.log10(Weinvsv[:, 0]), 'm:', label=r'$\sigma_1({We_1}^{-1})$')
plt.semilogx(omega, 20*np.log10(Weinvsv[:, 1]), 'g:', label=r'$\sigma_2({We_1}^{-1})$')
plt.ylabel('magnitude [dB]')
plt.xlim([1e-5, 1e3])
plt.xlabel('freq [rad/s]')
plt.legend()
plt.title('Singular values of inv(We)')
plt.ylim([-80, 20])
plt.fill_between(omega, -20, -200, where=(omega <= 0.1), color='c', alpha=0.2)

plt.figure(figsize=(15, 6))
# Plot singular values of inv(Wu)
plt.semilogx(omega, 20*np.log10(Wuinvsv[:, 0]), 'm:', label=r'$\sigma_1({Wu_1}^{-1})$')
plt.semilogx(omega, 20*np.log10(Wuinvsv[:, 1]), 'g:', label=r'$\sigma_2({Wu_1}^{-1})$')
plt.ylabel('magnitude [dB]')
plt.xlim([1e-5, 1e3])
plt.xlabel('freq [rad/s]')
plt.legend()
plt.title('Singular values of inv(Wu)')
plt.ylim([-40, 20])
plt.fill_between(omega, -10, -200, where=(omega > 200), color='c', alpha=0.2)
```

Generate the $G$ block of the LFT for the robust performance problem

```python
G = ct.minreal(ct.interconnect([P21, P12, K1, K2, A1, A3, W2,
                                W4, sump2, sump4, P11, P22, sumy11, sumy22,
                                sumy1, sumy2, We, Wu, fbk, sumnoise],
                    inputs=['wdelta[0]','wdelta[1]','wdelta[2]','wdelta[3]','r[0]','r[1]','n[0]','n[1]','u[0]','u[1]'],
                    outputs=['zdelta[0]','zdelta[1]','zdelta[2]','zdelta[3]','z[0]','z[1]','z[2]','z[3]','e[0]','e[1]']), 
                    verbose=False)
```

Do $\mu$-synthesis

```python
# Controller I/O sizes
f = 2

# Uncertainty structure: four 1x1 uncertainty blocks and a 4x4 performance block
nblock = np.array([1,1,1,1,4])
# This has to be == 2 (complex uncertainty) for each block (other values are not implemented)
itype = np.array([2,2,2,2,2])

# Frequency range for mu computations
omega = np.logspace(-3, 3, 121)

# Do mu-synthesis via D-K iteration
K, best_nubar, init_mubar, best_mubar, gamma = rb.musyn(G, f, nblock, itype, omega, reduce=16)

display("Best upper bound (nubar) to mu norm of Tzw_delta: ", best_nubar)
```

Plot initial and final values of $\mu$ upper bound

```python
plt.loglog(omega, best_mubar, label=r'Final $\bar \mu$')
plt.loglog(omega, init_mubar, label=r'Initial $\bar \mu$')
plt.ylabel(r'$\bar\mu$ value')
plt.xlabel('freq [rad/s]')
plt.legend()
plt.title(r'Initial and final $\bar\mu$')
```

Controller order reduction: perform a "balanced truncation" of the controller by removing almost uncontrollable or unobservable modes. Try reducing the order as long as the frequency response of the full and reduced controllers remain reasonably close. Then, evaluate the actual $\bar\nu$ achieved by the reduced controller.

```python
# Plot the frequency response of the controller
Ksv = rb.sigma(K, omega)
plt.semilogx(omega, 20*np.log10(Ksv[:, 0]), 'm', label=r'$\sigma_1(full)$')
plt.semilogx(omega, 20*np.log10(Ksv[:, 1]), 'g', label=r'$\sigma_2(full)$')

# Perform controller order reduction
order = 8
Kr = ct.balred(K, order, method='truncate')

# Plot the frequency response of the reduced controller
Krsv = rb.sigma(Kr, omega)
plt.semilogx(omega, 20*np.log10(Krsv[:, 0]), 'm:', label=r'$\sigma_1(red)$')
plt.semilogx(omega, 20*np.log10(Krsv[:, 1]), 'g:', label=r'$\sigma_2(red)$')
plt.ylabel('magnitude [dB]')
plt.xlabel('freq [rad/s]')
plt.legend()
plt.title('Comparison of singular values of full and reduced $K$')
```

```python
# After controller reduction, compute the actual nubar achieved by the reduced controller
# Generate the closed loop M transfer function for mu analysis
Kr.input_labels = ['e[0]','e[1]']
Kr.output_labels = ['u[0]','u[1]'] 
M = ct.minreal(ct.interconnect([P21, P12, K1, K2, A1, A3, W2,
                                W4, sump2, sump4, P11, P22, sumy11, sumy22,
                                sumy1, sumy2, We, Wu, fbk, sumnoise, Kr],
                    inputs=['wdelta[0]','wdelta[1]','wdelta[2]','wdelta[3]','r[0]','r[1]','n[0]','n[1]'],
                    outputs=['zdelta[0]','zdelta[1]','zdelta[2]','zdelta[3]','z[0]','z[1]','z[2]','z[3]'],
                    check_unused=False), 
                    verbose=False)

# Compute actual mu upper bound (frequency peak) for reduced controller
red_mubar, red_nubar = rb.mucomp(M, nblock, itype, omega)
display("Value of nubar with reduced controller:", red_nubar)

# Plot the mu upper bound for full and reduced controller
plt.loglog(omega, best_mubar, label=r'$\bar \mu$ (full)')
plt.loglog(omega, red_mubar, label=r'$\bar \mu$ (reduced)')
plt.ylabel(r'$\bar\mu$ value')
plt.xlabel('freq [rad/s]')
plt.legend()
plt.title(r'Full and reduced controller $\bar\mu$')
```

Verify performance on the nominal system.

```python
# Compute closed loop sensitivity for the nominal system
I = ct.ss([], [], [], np.eye(2)) # Identity system  

# Nominal plant
P = ct.tf([[[k10], [-0.05]],
              [[0.1], [-k20]]],
             [[[T10,1], [0.1,1]],
              [[0.3,1], [T20,1]]])


S = ct.minreal(ct.feedback(I,P*Kr))

# Frequency range for plots
omega = np.logspace(-5, 3, 121)

# Plot the singular value diagrams of S and KS, and the time simulation of the tracking error
plot_performance(omega, P, Kr, S, We, Wu, red_nubar)
```

Generate a number of plants in the uncertainty set and verify that the controller robustly satisfies the performance requirement

```python
plot_monte_carlo(Kr, We, Wu, omega, T1min, T1max, T2min, T2max, k1min, k1max, k2min, k2max, best_nubar)
```
